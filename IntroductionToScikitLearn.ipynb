# **Introduction to Scikit Learn**

It's used to handle machine learning. Machine learning are tools and techniques to understand data and predict what may happen in the future.

Scikit Learn for Linear Regression

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set() # Will plot the graphics with seaborn styles
import numpy as np
```

```python
n = 30
x = np.arange(n)
y = 4 * x - 7 + (10*np.random.randn(n))
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)
model.fit(x[:, np.newaxis], y)
yfit = model.predict(x[:,np.newaxis])
plt.scatter(x,y)
plt.plot(x, yfit, color="red")
```
