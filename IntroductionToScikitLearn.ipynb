# **Introduction to Scikit Learn**

It's used to handle machine learning. Machine learning are tools and techniques to understand data and predict what may happen in the future.

Scikit Learn for Linear Regression

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set() # Will plot the graphics with seaborn styles
import numpy as np
```

```python
n = 30
x = np.arange(n)
y = 4 * x - 7 + (10*np.random.randn(n))
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)
model.fit(x[:, np.newaxis], y)
yfit = model.predict(x[:,np.newaxis])
plt.scatter(x,y)
plt.plot(x, yfit, color="red")
```

Generating data with Scikit learn

```python
from sklearn.datasets._samples_generator import make_blobs

x, y = make_blobs(n_samples=500, centers=3, cluster_std=8)
plt.scatter(x[:,0], x[:,1], c=y, cmap="autumn")
```

Giving sample to train our model

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)
```

```python
from sklearn.svm import SVC
```

```python
mdl = SVC(gamma = 'auto')
mdl.fit(x_train, y_train)
```

```python
y_predict = mdl.predict(x_test)
```

Our accurary will be the calculation between our prediction made by our x_test, compared with out y_test and the test size.

```python
acc = 100 * np.sum(y_predict == y_test)/y_test.size
print("Accuracy is:", acc,"%")
```

Accurary will descrease along with our cluster_std increase

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
mdl_2 = RandomForestClassifier(n_estimators=20)
mdl_2.fit(x_train, y_train)
```

```python
y_predict_2 = mdl_2.predict(x_test)
acc_2 = 100 * np.sum(y_predict_2 == y_test)/y_test.size
print("Accuracy is:", acc_2,"%")
```
