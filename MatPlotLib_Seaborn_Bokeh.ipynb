# **MatPlotLib, Seaborn and Bokeh**

## Introduction to MatPlotLib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10,1000) # Creates 1000 values between 0 and 10
y = np.sin(x)
```

```python
plt.plot(x, y) # It will plot agraph of the values of X and its sin(x)
plt.plot(x[::200], y[::200], color="r")
plt.plot(x[::500], y[::500], color="g")
```

```python
plt.scatter(x[::20], y[::20], color="red")
```

## Seaborn

```python
# Plotting with MatPlotLib
plt.style.use('classic')
x_axis = np.linspace(0,100, 1000)
y_axis = np.cumsum(np.random.randn(1000,5), 0)
plt.plot(x_axis, y_axis)
plt.legend('ABCDE', ncol=2, loc="upper left")
```

```python
# Plotting with Seaborn
import seaborn as sns

sns.set()
plt.plot(x_axis, y_axis)
plt.legend('ABCDEF, ncol=2', loc="upper left")
```

```python
sns.set()
a = pd.Series(50*np.random.randn(5000))
b = pd.Series(200*np.random.randn(5000))
c = pd.Series(100*np.random.randn(5000) + 500)
data = pd.DataFrame({'A': a, 'B': b, 'C': c})
data.head()
for col in data.columns:
  plt.hist(data[col], density=True, alpha=0.5) # Alpha will set the opacity
  sns.kdeplot(data[col], shade=True)
```

```python
for col in data.columns:
  sns.distplot(data[col])
```
